```python
df['Bedrooms'] = df['Bedrooms'].astype(int)
```

```python
#There is an outlier detected in Rental price. Let's visualize it with box plot for even a better way and handle it
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(y=df['Rental_Price'], palette="Paired")
plt.title("Rental_Price Boxplot")
plt.show()
```

```python
Q1 = df["Rental_Price"].quantile(0.25)
Q3 = df["Rental_Price"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df["Rental_Price"] < lower_bound) | (df["Rental_Price"] > upper_bound)]
print(outliers)
```

```python
from scipy.stats.mstats import winsorize
df["Rental_Price"] = winsorize(df["Rental_Price"], limits=[0, 0.05])  
```
